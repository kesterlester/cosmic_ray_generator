```python
import numpy
import matplotlib.pyplot as plt
from scipy import optimize, integrate
import pandas
import sys
sys.path.insert(0,'/home/leo/Documents/cosmic_ray_generator/')
import functions
```

```python
def scaledTheory(t,E):
    scale=functions.composite(E_mu=E)/functions.theory_supressed(E_mu=E,theta=0)
    return scale*functions.theory_supressed(theta=t,E_mu=E)
```

```python
integrate.quad(scaledTheory, a=0, b=numpy.pi/2, args=(1,))
```

```python
x=numpy.linspace(0,numpy.pi/2,10000)
plt.plot(x,scaledTheory(x,1), label='E=1GeV')
```

```python
x=numpy.linspace(0.1,300,1000)
y=[integrate.quad(scaledTheory, a=0, b=numpy.pi/2,args=(i,)) for i in x]
plt.plot(x,y, label='integrated over angles')
plt.xscale('log')
plt.yscale('log')
plt.legend()
```
